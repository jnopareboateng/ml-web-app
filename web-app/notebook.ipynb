```python
## import the needed libraries

import pandas as pd
import numpy as np

## summary of first 10 rows of dataset
ufos = pd.read_csv('../1-Web-App/data/ufos.csv')
ufos.head(20)
```

```python
## feature engineering- making a dataframe from the selected features

ufos = pd.DataFrame({'Seconds': ufos['duration (seconds)'],
                      'Country': ufos['country'],
                      'Latitude': ufos['latitude'],
                      'Longitude': ufos['longitude']})

ufos.Country.unique()
```

```python
## remove missing data and normalize the df(only include sightings from 1 to 60 seconds)

ufos.dropna(inplace=True)

ufos = ufos[(ufos['Seconds'] >= 1) & (ufos['Seconds'] <= 60)] # normalize the data to include time from 1 to 60 seconds

ufos.info()
```

```python
## Data pre processing with label encoder(alphabetical)

from sklearn.preprocessing import LabelEncoder

ufos['Country'] = LabelEncoder().fit_transform(ufos['Country'])

ufos.head(10)
```

```python
## Building the model

from sklearn.model_selection import train_test_split

Selected_features = ufos[['Seconds','Latitude','Longitude']]

X = Selected_features
y = ufos['Country']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
## getting the classification report

from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

print(classification_report(y_test, predictions))
print('Predicted labels: ', predictions)
print('Accuracy: ', accuracy_score(y_test, predictions))
```

```python
## import pickle to flatten the model for use in a web application.

import pickle
model_filename = 'ufo-model.pkl'
pickle.dump(model, open(model_filename,'wb'))

model = pickle.load(open('ufo-model.pkl','rb'))
print(model.predict([[50,44,-12]]))
```

```python
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType

initial_type = [('float_input', FloatTensorType([None,4]))]

options = {id(model): {'nocl': True, 'zipmap':False}}
```

```python
onx = convert_sklearn(model, initial_types=initial_type, options=options)
with open("./model.onnx", "wb") as f:
    f.write(onx.SerializeToString())
```
